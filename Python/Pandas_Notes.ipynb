# PADNDAS

## A panda series is like a column in table. It is a one-dimensional array holding data of any type.

```python
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
print(myvar[1])
```

##SERIES IN PANDAS

###The followig is an example of a defult table

```python
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)
print(myvar[0])
```

### The following data frams have labels

```python
import pandas as pd
a = [1, 7, 2]
lable = ["x", "y", "z"]
myvar = pd.Series(data = a, index = lable)
print(myvar)
```

```python
import pandas as pd
calories = { "day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

##RENAMING COLUMNS

```python
data = {
    "age": [50, 40, 60],
    "qualified": [True, False, False]
}
idx = ["Sally", "Mary", "John"]
emps = pd.DataFrame(data, index=idx)
```

```python
print(emps)
```

```python
news_emps = emps.rename({"Sally": "Peter", "Mary": "Patrick", "John":"Paula" })
```

```python
print(emps)
```

```python
print(news_emps)
```
